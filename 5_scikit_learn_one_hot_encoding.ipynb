With Sklearn OneHotEcoder

When sparse=True, the encoded output is returned as a sparse matrix, while sparse=False returns a dense matrix.

```python
!pip install --upgrade scikit-learn
# Upgrade scikit-learn to the latest version. This will ensure the 'sparse' parameter is available for OneHotEncoder.

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', dtype=int, sparse_output=False)
# Replace 'sparse' with 'sparse_output' in the OneHotEncoder initialization.
# The 'sparse_output' parameter controls the output format in newer versions of scikit-learn.
```

```python
df_trf = ohe.fit_transform(df[['buyer','fruits','gender']])
```

```python
df_trf
```

```python
cols = ohe.get_feature_names_out()
```
